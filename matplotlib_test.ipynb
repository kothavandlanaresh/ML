```python
import matplotlib.pyplot as plt

import numpy as np

import pandas as pd
```

```python
x=[0,1,2,3,4,9]
y=[0,4,5,6,8,10]


#resize your graph

plt.figure(figsize=(5,3), dpi=150)
#plt.plot(x,y, label = '2x', color = 'red', marker = '*', linestyle = '--')

#plt.plot(x,y, 'r*--',label = '2x')
#shorthand notation
plt.plot(x,y, 'b^--',label = '2x')


x2=np.arange(0,4.5,0.5)
print(x2)

plt.plot(x2,x2**2,'r*:',label = 'x^2')


plt.title('our first graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()

plt.savefig('mygraph.png', dpi = 150)

plt.show()
```

line graph

```python
gas=pd.read_csv('gas_prices.csv')

gas.head()

plt.title('Line graph')
plt.xlabel('Year')
plt.ylabel('Prices')

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year,gas[country], marker = '.', label = country)


# plt.plot(gas.Year,gas.USA,'r*--', label = 'USA')
# plt.plot(gas.Year,gas.Canada,'b^:', label = 'Canada')

plt.xticks(gas.Year[::3])

plt.legend()

plt.savefig('gas price figure.png', dpi=300)


plt.show()
```
